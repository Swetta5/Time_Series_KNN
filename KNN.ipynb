```python
import pandas as pd
import numpy as np

# Correct File Path
file_path = r"C:\Users\CHARUSAT\Documents\Business_forecasting\New folder (6)\Purchased_Dataset.csv"

# Read the CSV File
df = pd.read_csv(file_path)

# Display Data
print(df.head())


x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

# Step 2: Load the Dataset

 df.head() displays the first five rows of the dataset.

# Step 3: Define Features (X) and Target Variable (Y)
Features (X) → Age & Estimated Salary.
 Target (y) → Whether the customer purchased the product (0 = No, 1 = Yes).



```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state =5)
knnclassifier = KNeighborsClassifier(n_neighbors=2)
knnclassifier.fit(x_train,y_train)
y_pred = knnclassifier.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

# Step 5: Train & Evaluate K-Nearest Neighbors (KNN) Classifier
 Trains a KNN model with k=5 (5 nearest neighbors).
 Predicts on test data (y_pred).
 Evaluates accuracy using metrics.accuracy_score().

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state =7)
knnclassifier = KNeighborsClassifier(n_neighbors=5)
knnclassifier.fit(x_train,y_train)
y_pred = knnclassifier.predict(x_test)
metrics.accuracy_score(y_test,y_pred)

```

# Cross Validation
Cross Validation is a technique which involves reserving a particular sample of a data set on which you do not train the model. Later, you test the model on this sample before finalizing the model.
Here are the steps involved in cross validation:
 You reserve a sample data set.
 Train the model using the remaining part of the data set.
 Use the reserve sample of the data set test (validation) set. This will help you to know the effectiveness of model performance. It your model delivers a positive result on validation data, go ahead with current model. It rocks!
